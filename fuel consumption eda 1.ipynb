<h2> Step 5: Emission V/S year</h2>

<h3>Steps for Visualisations</h3>

<ul>
  <li><strong>Data Selection:</strong> The 'EMISSIONS' column is used to calculate the average emissions for each year.</li>
  <li><strong>Styling and Color Palette:</strong> The plot is styled with 'whitegrid' and the 'Set2' color palette.</li>
  <li><strong>Line Plot:</strong> A line plot is created to show the trend of emissions over the years.</li>
  <li><strong>Plot Title and Labels:</strong> The plot is titled 'Trend of Emissions Over the Years' with labeled axes.</li>
  <li><strong>Customization:</strong> Tick labels are rotated, gridlines are added, and the lowest emissions year is annotated.</li>
     <li><strong>Data Grouping:</strong> The 'groupby' function is used to group the data by the 'YEAR' column. This allows us to analyze the emissions data based on each year.</li>
  <li><strong>Calculating Average Emissions:</strong> By applying the 'mean' function to the 'EMISSIONS' column within each group, we calculate the average emissions for each year. This provides a summarized view of the emissions data over time.</li>
</ul>

<p>The grouping and averaging of emissions data help us understand the overall trend and variation in emissions across different years. It allows us to identify specific years with the lowest or highest average emissions and provides insights into changes in emissions levels over time.</p>
<p>The resulting line plot effectively visualizes the emissions trend and highlights the year with the lowest average emissions.</p>

<p>The code also prints the lowest emissions year and value.</p>

<h3>Findings and Insights</h3>

<ul>
  <li>The line chart visualizes the trend of average emissions over the years.</li>
  <li>From 2002 to 2014, there is a declining trend in average emissions, indicating a potential improvement in emissions control measures.</li>
  <li>Notably, 2014 stands out as the year with the lowest average emissions, highlighting the effectiveness of emission reduction efforts during that period.</li>
  <li>After 2014, there is an upward trend in average emissions, suggesting a possible increase in emissions levels.</li>
  <li>In 2014-2015, there is a steep curve showing a significant increase in average emissions, with the value for 2015 reaching 245.</li>
  <li>Following the sharp increase, the trend continues with a gradual rise in average emissions.</li>
</ul>

<p>The analysis of average emissions over time helps in understanding the overall direction and variations in emissions levels. It provides insights into periods of improvement, such as the decline until 2014, and raises awareness of potential areas of concern, such as the subsequent increase in emissions. This information can be valuable for policymakers and stakeholders to evaluate the effectiveness of emission control strategies and guide future environmental initiatives.</p>

```python
# Group the data by 'YEAR' and calculate the average emissions for each year
average_emissions_by_year = df.groupby('YEAR')['EMISSIONS'].mean()

# Find the year with the lowest average emissions
lowest_emissions_year = average_emissions_by_year.idxmin()
lowest_emissions = average_emissions_by_year.min()

# Set style and color palette
sns.set(style="whitegrid")
sns.set_palette("Set2")

# Create the line plot
plt.figure(figsize=(12, 8))
sns.lineplot(data=average_emissions_by_year, marker='o', markersize=8, linewidth=2)

# Set plot title and labels
plt.title('Trend of Emissions Over the Years', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Emissions', fontsize=12)

# Customize tick labels and gridlines
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)

# Remove right and top spines
sns.despine()

# Annotate the lowest emissions year
plt.annotate(f'Lowest: {lowest_emissions:.2f}', xy=(lowest_emissions_year, lowest_emissions),
             xytext=(lowest_emissions_year-5, lowest_emissions+0.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# Show the plot
plt.show()

# Print the lowest emissions year and value
print(f"The year with the lowest average emissions is {lowest_emissions_year} with an average emissions value of {lowest_emissions:.2f}.")
```

<h2> Step 6: Car Models with most amount of Emission</h2>

<h3>Choosing a Threshold for Model Inclusion</h3>

<p>When selecting the top models based on emissions, we need to decide whether to include models with a low count of data points. In our analysis, we carefully considered this decision and here's why:</p>

<ol>
  <li><strong>Statistical Reliability:</strong> Models with only 1 or 2 data points may not give us a reliable estimate of their emissions. Having a small number of observations can lead to more variability and less confidence in the emissions values we calculate for those models.</li>
  <li><strong>Representativeness:</strong> Including models with a higher count of data points ensures that our findings are more representative of the broader population. We want to draw conclusions that are applicable to a larger set of similar models, and models with more data points give us a better understanding of their emissions performance.</li>
  <li><strong>Insights Quality:</strong> Models with a higher count of data points provide us with more robust insights and trends. The emissions data for these models are based on a larger number of observations, allowing us to have a more comprehensive understanding of their emissions behavior.</li>
</ol>

<p>Considering these factors, we decided to set a threshold of <strong>10</strong> data points as the minimum count for model inclusion. This threshold ensures that the top models we identify have a more significant presence in the data, increasing the reliability and generalizability of our findings.</p>

<h3>Insights and Findings</h3>
<ul>
  <li><strong>Model: 612 SCAGLIETTI</strong> has the highest average emissions of 418.50 g/km.</li>
  <li><strong>SAVANA 3500 PASSENGER</strong> and <strong>EXPRESS 3500 PASSENGER</strong> have the same average emissions of 388.59 g/km.</li>
  <li><strong>G 550</strong> has an average emissions of 386.83 g/km.</li>
  <li><strong>SAVANA 2500 PASSENGER</strong> has an average emissions of 381.37 g/km.</li>
</ul>
<p>These models exhibit relatively high average emissions, highlighting the need for measures to reduce emissions and mitigate environmental impact.</p>

```python
# Set the count threshold for model inclusion
count_threshold = 10

# Group the data by 'MODEL' and calculate the count and mean emissions for each model
model_stats = df.groupby('MODEL').agg({'EMISSIONS': ['count', 'mean']})

# Filter out models with count below the threshold
filtered_models = model_stats[model_stats[('EMISSIONS', 'count')] > count_threshold]

# Sort the filtered models based on mean emissions in descending order
top_models = filtered_models.sort_values(by=('EMISSIONS', 'mean'), ascending=False).head(5)

# Print the top models and their statistics
for model, (count, emissions_mean) in top_models.iterrows():
    print(f"Model: {model}")
    print(f"Count: {count}")
    print(f"Average Emissions: {emissions_mean:.2f} g/km")
    print()

```
